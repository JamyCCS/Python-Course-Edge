grouped by season [df.groupby('season)]. 

an object is created where multiple dataframe is created [can be accessed by df.groupby('season').get_group('2008'). here get_group is a aggregator]. 

df.groupby('season')['winner'] means only winner column is taken on group.

used df.groupby('season')['winner'].agg(lambda x: x.value_counts().idxmax()) for counting the max winner in a season

lambda x: x.value_counts().idxmax() means winners are counted and max valus's id is returned

```python
df.groupby('season')['winner'].agg(lambda x: x.value_counts().idxmax())
```
